```python
#set path:
path='/cbica/projects/wolf_satterthwaite_reward/Margaret/Day2/curation/code/iterations'
bids_path='/cbica/projects/wolf_satterthwaite_reward/Margaret/Day2/curation/BIDS'

iteration1='2'
prefix1='iter2'

iteration2='4'
prefix2='iter4'

#initialize bids layout
layout=BIDSLayout(bids_path)

#define df of validator outputs from iter 1.2 and 2
df_1=pd.read_csv(path+'/iteration'+iteration1+'/'+prefix1+'_summary.csv')
df_2=pd.read_csv(path+'/iteration'+iteration2+'/'+prefix2+'_summary.csv')
```
